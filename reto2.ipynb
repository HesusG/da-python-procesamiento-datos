# Tabla de Frequencia

```python
columns = ['estimated_diameter.meters.estimated_diameter_max','relative_velocity.kilometers_per_second']
n = len(df_neo_january)
k = 1 + int(np.log2(n))
for column in columns:
    cut_bins = pd.cut(df_neo_january[column], k)
    
    # Imprimir la cantidad de clases, el nombre de la columna y la tabla de contenidos
    print(f'Columna: {column}')
    print(f'Número de clases (k): {k}')
    print(f'Tabla de Frequencia:\n{cut_bins.value_counts()}')
    print('-' * 50)
```

```python
def frequency_table(df, columns):
    for column in columns:
        n = len(df)
        k = 1 + int(np.log2(n))
        cut_bins = pd.cut(df[column], k)
        print(f'Columna: {column}')
        print(f'Número de clases (k): {k}')
        print(f'Tabla de Frequencia:\n{cut_bins.value_counts().sort_index()}')
        print('-' * 50)
```

```python
frequency_table(df_nyt_bestsellers, ['price.numberDouble'])
```

```python
frequency_table(df_melbourne_housing, ['land_size'])
```

# Adicional: Histograma

```python
def my_histogram(df, columns):
    for column in columns:
        n = len(df)
        k = 1 + int(np.log2(n))
        plt.figure(figsize=(10, 6)) 
        sns.histplot(df[column], bins=k)
        plt.title(f'Histograma de {column}')
        plt.xlabel(column)
        plt.ylabel('Frecuencia')
        plt.tight_layout()  
        plt.show()
```

```python
my_histogram(df_nyt_bestsellers, ['price.numberDouble'])
my_histogram(df_melbourne_housing, ['land_size'])
```
