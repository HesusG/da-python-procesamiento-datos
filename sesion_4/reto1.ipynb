### a) Ruta 1: Near Earth Objects

El dataset 'near_earth_objects-jan_feb_1995-clean.csv' contiene datos acerca de objetos espaciales que orbitan cerca de la Tierra. Vamos a explorar las relaciones entre los tamaños y las velocidades de estos objetos.

```python
df_1 = pd.read_csv(ruta_1, index_col=0)
df_1.head()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    x=df_1['estimated_diameter.meters.estimated_diameter_min'], 
    y=df_1['estimated_diameter.meters.estimated_diameter_max']
)

plt.xlabel("Diámetro Estimado Mínimo (metros)")
plt.ylabel("Diámetro Estimado Máximo (metros)")
plt.title("Relación entre Diámetro Estimado Mínimo y Máximo de Asteroides")

correlation = df_1['estimated_diameter.meters.estimated_diameter_min'].corr(df_1['estimated_diameter.meters.estimated_diameter_max'])

plt.figtext(0.5, 0.01, f"Correlación de Pearson: {correlation:.2f}", ha="center", fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    x=df_1['estimated_diameter.meters.estimated_diameter_max'], 
    y=df_1['relative_velocity.kilometers_per_second']
)

plt.xlabel("Diámetro Estimado Máximo (metros)")
plt.ylabel("Velocidad Relativa (kilómetros por segundo)")
plt.title("Relación entre Diámetro Estimado Máximo y Velocidad Relativa de Asteroides")
correlation = df_1['estimated_diameter.meters.estimated_diameter_max'].corr(df_1['relative_velocity.kilometers_per_second'])
plt.figtext(0.5, 0.01, f"Correlación de Pearson: {correlation:.2f}", ha="center", fontsize=12)
plt.show()
```

### b) Ruta 2: Athlete Olympic Events

Este dataset contiene datos sobre eventos Olímpicos. La fuente original tiene información sobre el atleta que participó, las fechas y premios. Esta versión del dataset incluye solamente datos sobre la edad, peso y altura de los atletas. Comparemos edades, pesos y alturas.

Debido al elevado número de muestras en este dataset, utilizar gráficas de dispersión convencionales no resulta muy efectivo. Por esta razón, estas relaciones están hechas con una gráfica llamada *binning hexagonal*. Lo que hacen los binnings hexagonales es dividir todo el plano en pequeños hexágonos. Cada hexágono representa un rango de valores de `x` y `y`. Después, se cuenta el número de muestras que hay sobre cada hexágono y eso se asigna como la 'densidad' del hexágono. Cada hexágono se colorea dependiendo de la densidad: entre más obscuro, más puntos cayeron sobre ese hexágono.

```python
df_2 = pd.read_csv(ruta_2, index_col=0)
df_2.head()
```

```python
joint_plot = sns.jointplot(
    x=df_2['age'], 
    y=df_2['weight'], 
    kind='hex',
    height=10,
    marginal_kws={'fill': True}
)

joint_plot.ax_joint.set_xlabel("Edad")
joint_plot.ax_joint.set_ylabel("Peso")
joint_plot.fig.suptitle("Distribución Conjunta de Edad y Peso", fontsize=16, y=1.05)
plt.show()
```

El joinplot se distingue del scatterplot tradicional en varios aspectos clave, ofreciendo ventajas únicas, especialmente en el análisis de grandes conjuntos de datos:


- **Visualización de Densidades en Grandes Conjuntos de Datos**: Mientras un scatterplot muestra puntos individuales, un jointplot con hexágonos agrupa los puntos en hexágonos, con colores que representan la densidad de puntos. Esto es particularmente útil cuando hay una gran cantidad de datos, donde un scatterplot normal puede resultar en una sobreposición excesiva de puntos, haciendo difícil identificar patrones.

- **Reducción de la Superposición y Mejora de la Claridad**: Los hexágonos minimizan el problema de superposición y permiten una visualización más clara de la concentración de los datos, lo cual es difícil de discernir en un scatterplot cuando los puntos se superponen significativamente.

- **Interpretación Intuitiva de las Áreas de Alta Densidad**: Los colores más oscuros en los hexágonos indican áreas de mayor densidad de datos, lo que proporciona una interpretación visual intuitiva de dónde se concentran los datos.


### Cuándo Usar Jointplot con Hexágonos

- **Análisis de Grandes Conjuntos de Datos**: Es ideal para conjuntos de datos donde un scatterplot tradicional tendría demasiados puntos superpuestos, lo que puede hacer que el análisis sea confuso o menos informativo.

- **Identificación de Patrones en Datos Densos**: Cuando es importante entender no solo la relación entre dos variables, sino también identificar las áreas de mayor y menor concentración de datos.


```python

hex_plot = sns.jointplot(
    x='height', 
    y='weight', 
    data=df_2, 
    kind='hex',
    height=4, 
    marginal_kws={'fill': True},
)
hex_plot.set_axis_labels("Altura", "Peso")
hex_plot.fig.suptitle("Distribución Conjunta de Altura y Peso con Hexágonos")

scatter_plot = sns.jointplot(
    x='height', 
    y='weight', 
    data=df_2, 
    kind='scatter',
    height=4, 
    marginal_kws={'fill': True}
)
scatter_plot.set_axis_labels("Altura", "Peso")
scatter_plot.fig.suptitle("Distribución Conjunta de Altura y Peso con Scatterplot")
plt.tight_layout()
plt.show()
```

### c) Ruta 4: Ventas de videojuegos a través del mundo.

Este dataset contiene datos sobre ventas de videojuegos en todo el mundo. Vamos a ver las relaciones que existen entre las ventas en diferentes lugares del mundo.

```python
df_4 = pd.read_csv(ruta_3)
df_4.head()
```

```python
df_4 = df_4[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
```

```python
df_4 = df_4[df_4 > 0]
df_4 = df_4[df_4['NA_Sales'] < 20]
df_4 = df_4[df_4['JP_Sales'] < 10]
```

```python
plt.figure(figsize=(10, 6))

scatter_plot = sns.scatterplot(x=df_4['NA_Sales'], y=df_4['JP_Sales'])
plt.xlabel("Ventas en Norteamérica (millones)")
plt.ylabel("Ventas en Japón (millones)")
plt.title("Relación entre Ventas en Norteamérica y Japón")
correlation = df_4['NA_Sales'].corr(df_4['JP_Sales'])
plt.figtext(0.5, 0.01, f"Correlación de Pearson: {correlation:.2f}", ha="center", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x=df_4['NA_Sales'], y=df_4['JP_Sales'])
plt.xlabel("Ventas en Norteamérica (millones)")
plt.ylabel("Ventas en Japón (millones)")
plt.title("Relación entre Ventas en Norteamérica y Japón")
correlation = df_4['NA_Sales'].corr(df_4['JP_Sales'])
plt.figtext(0.5, 0.01, f"Correlación de Pearson: {correlation:.2f}", ha="center", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

scatter_plot = sns.scatterplot(x=df_4['NA_Sales'], y=df_4['Global_Sales'])
plt.xlabel("Ventas en Norteamérica (millones)")
plt.ylabel("Ventas Globales (millones)")
plt.title("Relación entre Ventas en Norteamérica y Ventas Globales")
correlation = df_4['NA_Sales'].corr(df_4['Global_Sales'])
plt.figtext(0.5, 0.01, f"Correlación de Pearson: {correlation:.2f}", ha="center", fontsize=12)
plt.show()
```
