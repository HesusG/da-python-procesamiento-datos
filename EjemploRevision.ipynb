## **Introducción**

#### La Importancia de la Seguridad Cibernética en la Era Digital

**En la era digital actual, la información se ha convertido en un activo valioso tanto para empresas como para individuos.** Sin embargo, esta creciente dependencia en el mundo digital ha desencadenado un aumento en los ataques cibernéticos, los cuales pueden acarrear consecuencias significativas, como la pérdida de datos confidenciales, interrupciones en los servicios y daños a la reputación.

#### Evolución de los Ataques Cibernéticos

Desde su aparición en 1988 con el **gusano de Morris**, que causó daños considerables a las computadoras y su conexión a internet, los ataques cibernéticos han evolucionado drásticamente en términos de **peligrosidad** y **frecuencia**. Tan solo en el año **2023**, se registraron más de **800,000 ciberataques**, lo que equivale a un promedio de **2,220 ataques diarios** (Charlton, 2024).

#### Impacto a Nivel Individual y Organizacional

- **Individual:** El caso de Joshua Schulte, condenado a 40 años de prisión por robar información clasificada de la CIA, ejemplifica el impacto personal de los ataques cibernéticos.
  
- **Organizacional:** El informe "Costo de una Brecha de Datos" de IBM (2023) revela que al sector público le cuesta en promedio 2.07 millones de dólares enfrentar estos ataques, mostrando las consecuencias financieras para las organizaciones.

#### Desafíos del Crecimiento Exponencial de la Información

El **crecimiento exponencial de la información** en el panorama digital actual plantea desafíos significativos en términos de **seguridad cibernética**. En **2020**, se generaron **64.2 zettabytes** de datos a nivel mundial, lo que representa una cantidad masiva de información digital. Se proyecta que esta cifra supere los **180 zettabytes para 2025** (Statista, 2023). Este aumento en la generación de datos subraya la urgente necesidad de desarrollar **protocolos eficientes y sólidos** para gestionar la información de manera segura, dado que está constantemente expuesta a diversos **ataques cibernéticos**.

#### Colaboración y Responsabilidad Colectiva

Es importante que todos los actores, desde individuos hasta grandes empresas y entidades gubernamentales, estén preparados para enfrentar los desafíos de la seguridad cibernética.

- **Consecuencias:** De lo contrario, podrían enfrentar pérdidas financieras, daños a la reputación y responsabilidades legales.
  
- **Mantenimiento y Actualización:** Por ende, es fundamental garantizar el mantenimiento y la actualización de los protocolos de seguridad, así como seguir las prácticas recomendadas para el cuidado de los datos informáticos, con el fin de mitigar los riesgos asociados con los ataques cibernéticos.

Es relevante comprender que la seguridad en línea es una **responsabilidad compartida** entre los usuarios y los proveedores de servicios. Tanto las **empresas privadas** como las **entidades gubernamentales** deben asegurar la **actualización** y el **mantenimiento de los protocolos de seguridad**. Los **usuarios**, por su parte, deben seguir **prácticas recomendadas para proteger sus datos informáticos**. Esta colaboración nos ayuda para mantener un **entorno digital seguro y protegido**.

```python
# Paso 12: Cálculo de Estimados de Locación para Columnas Numéricas
# ----------------------------------------------------------------
# Definimos una lista de las columnas numéricas en el DataFrame y calculamos los estimados de locación para estas columnas.

# Lista de nombres de columnas numéricas que queremos analizar.
columnas_numericas = ['Source Port', 'Destination Port', 'Packet Length', 'Anomaly Scores']

# Calculamos la media y la mediana de los valores en estas columnas.
# - La media representa el valor promedio de los datos
# - La mediana es el valor que separa la mitad superior e inferior de los datos ordenados.

media_valores = df_cyberattacks[columnas_numericas].mean()  # Calculamos la media para cada columna numérica.
mediana_valores = df_cyberattacks[columnas_numericas].median()  # Calculamos la mediana para cada columna numérica.
```

```python
# Paso 13: Cálculo de Estimados de Variabilidad para Columnas Numéricas
# ---------------------------------------------------------------------
# Definimos una lista de las columnas numéricas en el DataFrame y calculamos los estimados de variabilidad para estas columnas.

# Lista de nombres de columnas numéricas que queremos analizar.
columnas_numericas = ['Source Port', 'Destination Port', 'Packet Length', 'Anomaly Scores']

# Calculamos diferentes estimados de variabilidad para estas columnas.
# - El rango es la diferencia entre el valor máximo y mínimo en los datos.
# - La desviación estándar es una medida de la dispersión de los datos alrededor de la media.
# - El rango intercuartílico (IQR) es la diferencia entre el tercer y primer cuartil.

# Calculamos el rango, la desviación estándar y el IQR para cada columna numérica.
rango_valores = df_cyberattacks[columnas_numericas].max() - df_cyberattacks[columnas_numericas].min()  # Calculamos el rango.
desviacion_estandar_valores = df_cyberattacks[columnas_numericas].std()  # Calculamos la desviación estándar.
iqr_valores = df_cyberattacks[columnas_numericas].quantile(0.75) - df_cyberattacks[columnas_numericas].quantile(0.25)  # Calculamos el IQR.
```

```python
# Paso 14: Creación de un DataFrame de Resultados
# ----------------------------------------------
# Creamos un DataFrame llamado 'resultados' que contiene los estimados de locación y variabilidad para las columnas numéricas.

# Creamos el DataFrame 'resultados' con las siguientes columnas:
# - 'Media': contiene los valores de media calculados previamente.
# - 'Mediana': contiene los valores de mediana calculados previamente.
# - 'Rango': contiene los valores de rango calculados previamente.
# - 'Desviación Estándar': contiene los valores de desviación estándar calculados previamente.
# - 'IQR': contiene los valores de rango intercuartil (IQR) calculados previamente.

resultados = pd.DataFrame({
    'Media': media_valores,
    'Mediana': mediana_valores,
    'Rango': rango_valores,
    'Desviación Estándar': desviacion_estandar_valores,
    'IQR': iqr_valores
})

# Mostramos el DataFrame 'resultados' que resume los estimados de locación y variabilidad.
print(resultados)
```
