```python
# Paso 12: Clculo de Estimados de Locacin para Columnas Numricas
# ----------------------------------------------------------------
# Definimos una lista de las columnas numricas en el DataFrame y calculamos los estimados de locacin para estas columnas.

# Lista de nombres de columnas numricas que queremos analizar.
columnas_numericas = ['Source Port', 'Destination Port', 'Packet Length', 'Anomaly Scores']

# Calculamos la media y la mediana de los valores en estas columnas.
# - La media representa el valor promedio de los datos
# - La mediana es el valor que separa la mitad superior e inferior de los datos ordenados.

media_valores = df_cyberattacks[columnas_numericas].mean()  # Calculamos la media para cada columna numrica.
mediana_valores = df_cyberattacks[columnas_numericas].median()  # Calculamos la mediana para cada columna numrica.
```

```python
# Paso 13: Clculo de Estimados de Variabilidad para Columnas Numricas
# ---------------------------------------------------------------------
# Definimos una lista de las columnas numricas en el DataFrame y calculamos los estimados de variabilidad para estas columnas.

# Lista de nombres de columnas numricas que queremos analizar.
columnas_numericas = ['Source Port', 'Destination Port', 'Packet Length', 'Anomaly Scores']

# Calculamos diferentes estimados de variabilidad para estas columnas.
# - El rango es la diferencia entre el valor mximo y mnimo en los datos.
# - La desviacin estndar es una medida de la dispersin de los datos alrededor de la media.
# - El rango intercuartlico (IQR) es la diferencia entre el tercer y primer cuartil.

# Calculamos el rango, la desviacin estndar y el IQR para cada columna numrica.
rango_valores = df_cyberattacks[columnas_numericas].max() - df_cyberattacks[columnas_numericas].min()  # Calculamos el rango.
desviacion_estandar_valores = df_cyberattacks[columnas_numericas].std()  # Calculamos la desviacin estndar.
iqr_valores = df_cyberattacks[columnas_numericas].quantile(0.75) - df_cyberattacks[columnas_numericas].quantile(0.25)  # Calculamos el IQR.
```

```python
# Paso 14: Creacin de un DataFrame de Resultados
# ----------------------------------------------
# Creamos un DataFrame llamado 'resultados' que contiene los estimados de locacin y variabilidad para las columnas numricas.

# Creamos el DataFrame 'resultados' con las siguientes columnas:
# - 'Media': contiene los valores de media calculados previamente.
# - 'Mediana': contiene los valores de mediana calculados previamente.
# - 'Rango': contiene los valores de rango calculados previamente.
# - 'Desviacin Estndar': contiene los valores de desviacin estndar calculados previamente.
# - 'IQR': contiene los valores de rango intercuartil (IQR) calculados previamente.

resultados = pd.DataFrame({
    'Media': media_valores,
    'Mediana': mediana_valores,
    'Rango': rango_valores,
    'Desviacin Estndar': desviacion_estandar_valores,
    'IQR': iqr_valores
})

# Mostramos el DataFrame 'resultados' que resume los estimados de locacin y variabilidad.
print(resultados)
```
